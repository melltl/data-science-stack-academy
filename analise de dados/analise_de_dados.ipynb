#merge entre df diferentes

```python
orders= pd.read_csv('olist_orders_dataset.csv')
orders.head()
```

```python
orders_items= pd.read_csv('olist_order_items_dataset.csv')
orders_items.head()
```

##opcoes de merge
`inner join: retorna apenas os pedidos que contem a msm chave`



`left join: retorna todos os registros a esquerda`



`right join: retorna todos os registros adireita`



`outer join: retorna todos os registros`

```python
query=pd.merge(orders[['order_id','order_status',"order_approved_at"]],
               orders_items[['order_id','product_id','seller_id', 'price']],
               on='order_id')
query.head()
```

```python
query.isnull().sum()
```

###left join

```python
query=pd.merge(orders[['order_id','order_status',"order_approved_at"]],
               orders_items[['order_id','product_id','seller_id', 'price']],
               on='order_id',
               how='left')
query.head()
```

```python
query.isnull().sum()
```

###right join

```python
query=pd.merge(orders[['order_id','order_status',"order_approved_at"]],
               orders_items[['order_id','product_id','seller_id', 'price']],
               on='order_id',
               how='right')
query.head()
```

```python
query.isnull().sum()
```

##outer

```python
query=pd.merge(orders[['order_id','order_status',"order_approved_at"]],
               orders_items[['order_id','product_id','seller_id', 'price']],
               on='order_id',
               how='outer')
query.head()
```

```python
query.isnull().sum()
```
